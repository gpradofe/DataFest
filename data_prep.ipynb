```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data into a pandas dataframe
attorneys = pd.read_csv('attorneys.csv')
attorney_time_entries = pd.read_csv('attorneytimeentries.csv')
categories = pd.read_csv('categories.csv')
questions = pd.read_csv('questions.csv')

# Join the data to get the number of questions by category for each attorney
attorney_question_counts = pd.merge(
    questions[['TakenByAttorneyUno', 'Category']],
    attorney_time_entries[['AttorneyUno', 'TimeEntryUno']],
    left_on='TakenByAttorneyUno',
    right_on='AttorneyUno'
)
```

```python
new = attorney_question_counts.merge(
    categories[['CategoryUno', 'Category']],
    on='Category'
)
```
